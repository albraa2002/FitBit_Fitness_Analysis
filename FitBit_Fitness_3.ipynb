```python
workout = minIntN.loc[minIntN['Intensity'] >= 2]
workout.head()
```

```python
workout.shape
```

```python
len(workout['Id'].unique())
```

```python
workout['dt-datetime'] = workout['ActivityMinute'].map(lambda x: pd.to_datetime(x))
workout.head()
```

```python
workout['dt-date'] = workout['dt-datetime'].map(lambda x: x.date())
workout.head()
```

```python
workout['dt-time'] = workout['dt-datetime'].map(lambda x: x.time())
workout.head()
```

```python
dates = workout['dt-date'].unique()
dates
```

```python
Ids = workout['Id'].unique()
Ids
```

```python
times = workout[['Id', 'dt-datetime']]
times
```

```python
wko_start_times = []
wko_end_times = []
wko_ids = []
wko_dates = []
```

```python
start_time = times.iloc[0]['dt-datetime']
start_time
```

```python
wko_start_times.append(start_time)
wko_start_times
```

```python
wko_ids.append(times.iloc[0]['Id'])
wko_dates.append(str(start_time)[:10])
```

```python
type(start_time)
```

```python
d = str(start_time)[:10]
d
```

```python
for user in Ids: 
    one_user = times.loc[times['Id'] == user][['Id', 'dt-datetime']]
    for i in list(range(1, len(one_user))):
        t = one_user.iloc[i]['dt-datetime']
        delta = t - one_user.iloc[i-1]['dt-datetime']
        if (delta.seconds >= 1800) & (delta.days < 1):
            end_time = one_user.iloc[i-1]['dt-datetime']
            start_time = t
            wko_start_times.append(start_time)
            wko_end_times.append(end_time)
            wko_ids.append(user)
            wko_dates.append(str(start_time)[:10])
```

```python
wko_end_times.append(times.iloc[-1]['dt-datetime'])
```

```python
print(len(wko_start_times))
print(len(wko_end_times))
print(len(wko_ids))
print(len(wko_dates))
```

```python
dict_for_df = {'Id': wko_ids,
                'Date': wko_dates,
                'StartTime': wko_start_times,
                'EndTime': wko_end_times}
workout_tracktime = pd.DataFrame(dict_for_df)
workout_tracktime.head()
```

```python
workout_tracktime['duration'] = workout_tracktime['EndTime'] - workout_tracktime['StartTime']
workout_tracktime.head()
```

```python
workout_tracktime['duration'].describe()
```

```python
duration_outliers = workout_tracktime.loc[workout_tracktime['duration'] > pd.Timedelta(days=1)]
duration_outliers.describe()
```
